```python
import subprocess as sp
import numpy as np
import matplotlib.pyplot as plt

FFMPEG_BIN = "ffmpeg\\bin\\ffmpeg.exe"

vfile = 'UCF-101\\Archery\\v_Archery_g01_c01.avi'
command = [FFMPEG_BIN,
           '-i', vfile,
           '-f', 'image2pipe',
           '-pix_fmt', 'rgb24',
           '-vcodec', 'rawvideo', '-']
pipe = sp.Popen(command, stdout = sp.PIPE, bufsize = 10 * 1024 * 1024)

for i in range(0, 90):
    raw_image = pipe.stdout.read(320*240*3)
    if i % 10 == 0:
        image = np.fromstring(raw_image, dtype='uint8')
        image = image.reshape((240, 320, 3))
        print(image.shape)
        plt.imshow(image)
        plt.show()

pipe.stdout.flush()
```

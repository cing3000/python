```python
from IPython.display import clear_output
from urllib.request import Request
import urllib.request
import json
import matplotlib.pyplot as plt
import numpy as np
import time
import os
import re
from datetime import datetime, timedelta, timezone
from six.moves import cPickle as pickle

# initial http proxy
proxy = urllib.request.ProxyHandler({'https': 'https://dzlu:vD#j8n5t@podclx139-wcg.qdot.qld.gov.au:8080'})
auth = urllib.request.HTTPBasicAuthHandler()
opener = urllib.request.build_opener(proxy, auth, urllib.request.HTTPHandler)
urllib.request.install_opener(opener)

symbols = ['DOGEUSD', 'LTCUSD', 'XMRUSD', 'ETHUSD', 'DASHUSD', 'ZECUSD']
for symbol in symbols:
    req = Request('https://api.hitbtc.com/api/2/public/candles/%s?period=D1&limit=1000' % symbol)
    req.add_header('Accept', 'text/html,application/json,application/xml;q=0.9,image/webp,*/*;q=0.8')
    req.add_header('User-Agent', 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36')
    req.add_header('Upgrade-Insecure-Requests', '1')

    resp = urllib.request.urlopen(req)
    
    data = []
    raw_data = json.loads(resp.read().decode())
    
    # Remove first 10 days record
    if (len(raw_data) < 1000):
        raw_data = raw_data[10:]

    for row in raw_data:
        r_date = row['timestamp'][:10].replace('-', '')
        item = {
            'ticker': symbol,
            'timestamp': r_date,
            'open': row['open'],
            'high': row['max'],
            'low': row['min'],
            'close': row['close'],
            'vol': row['volume']
        }
        data.append(item)
    
    # Save it
    filename = symbol + ".pickle"
    try:
        with open(filename, 'wb') as f:
            pickle.dump(data, f, pickle.HIGHEST_PROTOCOL)
    except Exception as e:
        print('Unable to save data to', set_filename, ':', e)
```

```python
import os
import csv
import mysql.connector

cnx = mysql.connector.connect(user='root', host='127.0.0.1', database='dig')
cursor = cnx.cursor()

insert = ("INSERT INTO DATA (symbol, date, open, high, low, close, vol) values (%s, %s, %s, %s, %s, %s, %s)")

home_dir = "raw\\"
folders = os.listdir(home_dir)
c = 0

for folder in folders:
    files = os.listdir(home_dir + folder)
    for file in files:
        print(file)
        with open(home_dir + folder + "\\" + file, 'r') as csv_file:
            csv_file.readline()
            sheet = csv.reader(csv_file)
            for row in sheet:
                if (row[6] == '0'):
                    continue
                data = (row[0], row[1], row[2], row[3], row[4], row[5], row[6])
                cursor.execute(insert, data)
                c = c + 1
                if (c % 5000 == 0):
                    print(c)
                    cnx.commit()
        
cursor.close()
cnx.close()
```

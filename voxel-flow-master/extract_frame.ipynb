```python
import cv2
import os
from os import listdir
from os.path import isfile, join

ucf_folder = '..\\UCF-101'
count = 1
for folder in listdir(ucf_folder):
    for avi_file in listdir(join(ucf_folder, folder)):
        video_cap = cv2.VideoCapture(join(ucf_folder, folder, avi_file))
        length = int(video_cap.get(cv2.CAP_PROP_FRAME_COUNT))

        pos = 0
        while pos + 5 < length:
            
            target_dir = ".\\data\\%d" % ((count/500)+1)
            try:
                os.stat(target_dir)
            except:
                os.mkdir(target_dir)

            video_cap.set(1, pos) # CV_CAP_PROP_POS_FRAMES = 1
            _, image = video_cap.read()
            image = cv2.resize(image, (256, 256))
            cv2.imwrite("%s\\frame_%d_1.png" % (target_dir, count), image)
            video_cap.set(1, pos+2) # CV_CAP_PROP_POS_FRAMES = 1
            _, image = video_cap.read()
            image = cv2.resize(image, (256, 256))
            cv2.imwrite("%s\\frame_%d_2.png" % (target_dir, count), image)
            video_cap.set(1, pos+4) # CV_CAP_PROP_POS_FRAMES = 1
            _, image = video_cap.read()
            image = cv2.resize(image, (256, 256))
            cv2.imwrite("%s\\frame_%d_3.png" % (target_dir, count), image)
            count += 1
            pos += 8
        
        print("%d trilets captured" % count)

#vidcap = cv2.VideoCapture('..\\UCF-101\\Archery\\v_Archery_g01_c01.avi')
#success,image = vidcap.read()
#count = 0
#success = True
#while success:
#cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
#success,image = vidcap.read()
#print('Read a new frame: ', success)
#count += 1
```
